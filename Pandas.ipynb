```python
#concate
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
})

df2 = pd.DataFrame({
    'ID': [4, 5, 6],
    'Name': ['David', 'Eve', 'Frank'],
    'Score': [88, 92, 80]
})

# Concatenate DataFrames
concated_df = pd.concat([df1, df2], ignore_index=True)

# Sort by 'Score' in descending order
sorted_df = concated_df.sort_values(by='Score', ascending=False)

print(sorted_df)
```
