#Subplot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,15,5,30]

fig = plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar chart')

plt.tight_layout()
fig.suptitle("Comparison between line and bar chart", y=1.1)
plt.show()
```

#Netflix Project


### Steps to get the raw GitHub file URL:

1.  **Navigate to the file on GitHub:** Go to the repository and find the file you want to load.
2.  **Click the "Raw" button:** This button is usually located near the top right of the file's content view.
3.  **Copy the URL:** The page that opens will display the raw content of the file. Copy the URL from your browser's address bar. This is the URL you'll use in pandas.

```python
import pandas as pd

# Replace with your actual raw GitHub URL
github_file_url = 'https://raw.githubusercontent.com/ariyan378/Python-Library/main/netflix_titles.csv'
# For CSV files
df = pd.read_csv(github_file_url)

df.head()
```

#Data Cleaning


```python
df = df.dropna(subset=['type','country','release_year','rating','duration'])
```

```python
df.head()
```

```python

df.columns = df.columns.str.title()
```

```python
df
```

```python
df.cast= df.Cast.fillna('Unknown',inplace=True)
```

```python
df.head()
```

```python

df['Director'] = df['Director'].fillna(df['Director'].mode()[0])
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Date_Added'] = df['Date_Added'].fillna(df['Date_Added'].mode()[0])
```

```python
df.head()
```

```python
df['Description'] = df['Description'].str.title()
```

```python
df.head(5)
```

#Type Count Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

type_count = df['Type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_count.index , type_count.values, color=['purple', 'pink'], edgecolor='black')
plt.title('Number of Movies and Tv Shows on netflix')
plt.xlabel('Type')
plt.ylabel('Counts')
plt.tight_layout()

plt.savefig('Movie Vs Tv Show on Netflix.jpg', dpi = 200, bbox_inches='tight')
plt.show()

```

#Rating Count Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

rating_count = df['Rating'].value_counts()

plt.pie(rating_count , labels = rating_count.index , autopct = '%1.1f%%', startangle=90)

plt.title('Rating of Movies and Tv Shows on netflix')

plt.tight_layout()

plt.savefig('Contents Rating.png', dpi = 200, bbox_inches='tight')
plt.show()
```

#Movie Duration Histrogram


```python
import matplotlib.pyplot as plt

movie_df = df[df['Type'] == 'Movie'].copy()
movie_df.loc[:, 'duration_int'] = movie_df['Duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(8, 6))
plt.hist(movie_df['duration_int'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.xlim(0,250)
plt.grid(axis='y', alpha=0.75)

plt.tight_layout()

plt.savefig('Distribution of Duration of Movie.png', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

country_counts = df['Country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(country_counts.index, country_counts.values, color='skyblue', edgecolor='black') # Using barh for horizontal bars
plt.title('Top 10 Countries by Content Release')
plt.xlabel('Counts') # X-axis is now counts
plt.ylabel('Country') # Y-axis is now country
plt.grid(axis='x', alpha=0.50) # Grid on x-axis for horizontal bars

# Invert y-axis to have the highest count at the top
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Release By Country Top 10.png' , dpi=300,bbox_inches='tight')
plt.show()
```

```python
from typing import ContextManager
content_by_year = df.groupby(['Release_Year', 'Type']).size().unstack().fillna(0)

fig,ax = plt.subplots(1,2,figsize=(12,6))

#movies

ax[0].plot(content_by_year.index,content_by_year['Movie'], color='red')
ax[0].set_title("Movies Release Per Year")
ax[0].set_xlabel('Release Year')
ax[0].set_ylabel('Number Of Movie')

#tv Shows

ax[0].plot(content_by_year.index,content_by_year['TV Show'], color='blue')
ax[0].set_title("TV Shows Release Per Year")
ax[0].set_xlabel('Release Year')
ax[0].set_ylabel('Number Of Shows')

fig.suptitle('Comparison of Movies and TV Shows Release Over Years')
plt.tight_layout()


plt.savefig('Comparison of Movies and TV Shows Release Over Years.png' , dpi=300,bbox_inches='tight')
plt.show()
```
