DataFrame Basics


```python
# Let's create a sample DataFrame to work with
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 28],
    'Salary': [50000, 60000, 75000, 55000],
    'City': ['NYC', 'LA', 'Chicago', 'Boston']
})

# Now let's explore it!
print(df.shape)        # Output: (4, 4) - means 4 rows and 4 columns
print(df.columns)      # Shows: ['Name', 'Age', 'Salary', 'City']
print(df.index)        # Shows: RangeIndex(start=0, stop=4, step=1)
print(df.dtypes)       # Shows the data type of each column
df.info()              # Gives a nice summary: column names, types, memory usage
df.describe()          # Statistical summary for numerical columns
```

```python
bios['Height_Category'] = bios['height_cm'].apply( lambda x : 'Short' if x < 170 else ('Average ' if x < 185 else 'Tall'))
```

```python
bios
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 170 and row['weight_kg'] < 70 :
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 90 :
        return " Midweight"
    
    else :
        return 'Heavyweight'
    
bios ['Category'] = bios.apply(categorize_athlete , axis = 1) 

bios
```
